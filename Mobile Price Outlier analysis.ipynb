```python
##Outlier Analysis
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
col=['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g',
       'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height',
       'px_width', 'sc_h', 'sc_w', 'talk_time', 'three_g', 'touch_screen',
       'wifi', 'price_range']
for c in col:
    per25=df[c].quantile(0.25)
    per75=df[c].quantile(0.75)
    iqr=per75-per25
    upperlimit=per75+(1.5*iqr)
    lowerlimit=per25-(1.5*iqr)
    A=df[(df[c]>lowerlimit) & (df[c]<upperlimit)]
    plt.figure()
    sns.boxplot(y=c,data=A)
```
